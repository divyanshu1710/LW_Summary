```python
#importing the required libraries for the project
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#fetching the data from the system 
data = pd.read_csv("ParisHousing.csv")
data
```

```python
#filling the vacant spaces and filling it with the 0
data.fillna(0)
```

```python
#getting valueable insights from the dataset
data.describe()
```

```python
#getting the attributes name
data.columns
```

```python
#initialising the independent attributes in x to presdict result
x=data[["squareMeters","numPrevOwners","numberOfRooms","hasGuestRoom","hasPool","hasStorageRoom","hasStormProtector","floors","cityCode","cityPartRange","isNewBuilt","made","basement","garage"]]
x
```

```python
#dependent variable or target varibale
y= data["price"]
y
```

```python
#initialising the linear regression model
model= LinearRegression()
```

```python
#fitting the model across the x,y
model.fit(x,y)
```

```python
#checking weights of the attributes
model.coef_
```

```python
#checking bias
model.intercept_
```

```python
#getting our price predicted
model.predict([[60000,3,5,3,1,0,2,23,9796,3,1,2003,3333,976]])
```
